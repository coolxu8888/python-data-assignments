```python
#爬取建筑网观点类文章标题
import requests
from bs4 import BeautifulSoup
import csv

def scrape_onepage(purl):
    r = requests.get(purl).text
    data = BeautifulSoup(r,"html.parser")
    all_div = data.find('div',attrs={'class':"wrapper fixed"}).find_all("li")
    tittles=[]
    for i in all_div:
        tittle=[]
        tittle.append(i.find('div', attrs={'class':"t"}).find('strong').text.strip())
        tittle.append(i.find('div', attrs={'class':"w"}).text.strip())
        tittle.append(i.find('div', attrs={'class':"t"}).find('a')['href'])
        tittles.append(tittle)
    return tittles
def scrape_all_pages(url):
    page_urls=[]
    comments=[]
    for i in range(1,11):  
        if i == 1:
            url = 'http://www.chinabuildingcentre.com/list-41-1.html'
        else:
            url = 'http://www.chinabuildingcentre.com/list-41-{}.html'.format(i)
        page_urls.append(url)
    for purl in page_urls:
        comments.extend(scrape_onepage(purl))
    return(comments)
result = scrape_all_pages('http://www.chinabuildingcentre.com/list-41-1.html')


with open('cbc.csv','w', encoding = 'utf-8') as f:

    writer = csv.writer(f)
    header = ['titles','interviewee','url']
    writer.writerow(header)
    writer.writerows(result)

import pandas as pd
pd.read_csv('cbc.csv', encoding = 'utf-8')

ef = pd.read_csv('cbc.csv', encoding = 'utf-8')
ef#显示统计结果
```

```python
#提取文章标题中的关键字，做分类统计
title = ef['titles']
titless = []
for i in title:
    titles = []
    titles.append(i)
    titless.append(titles)
file=open('data1.txt','w', encoding = 'utf8') 
file.write(str(titless)); 
file.close()
#接下来查找title里面重要的词的词频,data1里无用的符号已用word的替换功能清理,清理后为data，由于中文分词要使用新module，故直接翻译成英文再进行分词
news = open('datatranslate.txt', 'r')
contents = news.read()
def keywordfrequency(contents):
    list1 = contents.split( )
    list_mw = []
    word_f ={}
    with open ('stopwords-Copy1.txt','r') as s:
        list_stop_words = s.read().split()

    for i in list1:
        if i not in list_stop_words:
            list_mw.append(i)
    for i in list_mw:
        word_f[i] =list_mw.count(i)

    dict_fr={}
    list_top = sorted(word_f.items(), key=lambda item: item[1], reverse=True)

    list_top1 = list_top[0:50]
    for m in range(0,len(list_top1)):
        dict_fr.update({list_top1[m][0]:list_top1[m][1]})
    list_csv = []
    for i in dict_fr.items():
        list_csv.append(i)
    return list_csv
keywordfrequency(contents)
import csv
with open('cnckeywords_frequency.csv','w') as f:
    writer = csv.writer(f)
    header = ['keyword','frequency']
    writer.writerow(header)
    writer.writerows(list_csv)

pd.read_csv('cnckeywords_frequency.csv')#标题中关键词统计结果
```
